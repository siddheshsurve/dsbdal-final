```python
df['Number'] = df['Number'].fillna(df['Number'].mean())
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df
```

```python
bins = [19, 25, 30, 36, 40]
labels = ['19-24', '25-29', '30-35', '36-40']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df
```

```python
df.groupby('AgeGroup')['Salary'].mean().describe()
```

```python
salaryList = list(df.groupby('AgeGroup')['Salary'])
salaryList
```

```python
crosstab = pd.crosstab(index = df['Salary'], values = df['Salary'], columns = df['AgeGroup'], aggfunc = ['min', 'max', 'mean', 'median', 'std'])
crosstab
```
