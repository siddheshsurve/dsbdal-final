```python
bins = [0.1, 1.0, 2.0, 3.0]
labels = ['0.1-1.0', '1.1-2.0', '2.1-3.0']
df['PetalWidthGroup'] = pd.cut(df['PetalWidthCm'], bins=bins, labels=labels)
df
```

```python
df.groupby('PetalWidthGroup')['Species'].describe()
```

```python
petalList = list(df.groupby('PetalWidthGroup')['Species'])
petalList
```
